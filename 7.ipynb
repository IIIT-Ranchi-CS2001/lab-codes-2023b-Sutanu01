```python
import random 

K=int(input("Enter K: "))
N=int(input("Enter N: "))
newL = [random.randint(1, N+1) for _ in range(K)]
print(newL)
def isPrime(a):
    for i in range(2, int(a**0.5) + 1):
        if a % i == 0:
            return False
    return True
def getPrimes(list1):
    return [[i for i in list1 if isPrime(i)], [i for i in list1 if not isPrime(i)]]
p1=getPrimes(newL)[0]
c1=getPrimes(newL)[1]

primeSq=[i**2 for i in p1]
compositeSq=[i**(1/2) for i in c1]

fig, axs = plt.subplots(1,2)
axs[0].scatter(p1, primeSq, color='r')
axs[0].set_xlabel('Prime Numbers')
axs[0].set_ylabel('Square of Prime Numbers')
axs[1].scatter(c1, compositeSq, color='b')
axs[1].set_xlabel('Composite Numbers')
axs[1].set_ylabel('Square Root of Composite Numbers')
fig.tight_layout()

```
